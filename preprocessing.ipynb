```python
import csv
import pandas as pd
from textblob import TextBlob
```

```python
filename = "raw/1429_1.csv"
```

```python
with open(filename) as f:
    reader = csv.reader(f)
    items = [x for x in reader]
```

```python
df = pd.read_csv(filename)
```

```python
df.head()
```

```python
columns = ['id', 'reviews.title', 'reviews.text', 'reviews.rating']
filtered = df[columns]
```

```python
filtered.head()
```

```python
titles = df['reviews.title']    
texts = df['reviews.text']
ratings = df['reviews.rating']
```

```python
title_blobs = [TextBlob(str(x)) for x in texts]
text_blobs = [TextBlob(str(x)) for x in titles]
```

```python
title_sentiments = [b.sentiment.polarity for b in title_blobs]
text_sentiments = [b.sentiment.polarity for b in text_blobs]
```

```python
predict_df = pd.DataFrame([(titles[i], texts[i], inratings[i]), title_sentiments[i], text_sentiments[i]) for i in range(len(titles))])
```

```python
predict_df.head()
```

```python
predict_df.to_csv("raw/sentiments.csv", header=True)
```
